```python
import pandas as pd

data = pd.read_csv('recipes.csv')

print(data)



```

```python
ingredient_list = data['ingredients']

for ingredients_str in ingredient_list:
    ingredients = ingredients_str.split(',')  # Store split ingredients in a new variable

    for ingredient_str in ingredients:
        ingredient_str = ingredient_str.strip()  # Corrected strip function
        print(ingredient_str)  
```

```python
import spacy
# Load the NLP model (ensure you have installed `en_core_web_sm`)
nlp = spacy.load("en_core_web_sm")

# Sample direction text
directions = "Chop the onions and garlic, then sauté them in butter until golden. Add diced tomatoes and let simmer."

# Process the text
doc = nlp(directions)

# Extract food-related entities
extracted_ingredients = [ent.text for ent in doc.ents if ent.label_ == "FOOD"]

print(extracted_ingredients)
```
